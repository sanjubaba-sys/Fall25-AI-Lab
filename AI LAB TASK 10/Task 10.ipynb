# Data Pre-Processing

```python
import pandas as pd
import numpy as np

data = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
```

#  print Rows aur Columns 

```python
print('We have {} rows.'.format(data.shape[0]))
print('We have {} columns'.format(data.shape[1]))
```

```python
np.sum(pd.isnull(data))
```

```python
data['fuel'].unique()
```

```python
num = data['fuel'].mode()[0]
data['fuel'] = data['fuel'].fillna(num)
```

```python
data['year'] = data['year'].astype(np.int64)
```

```python
data.dtypes
```

```python
x = data.iloc[:, 0:-1]
y = data.iloc[:, -1]

print(x.shape)
print(y.shape)
```

```python
cat_columns = x.select_dtypes(['object']).columns
x[cat_columns] = x[cat_columns].apply(lambda x: pd.factorize(x)[0])
```
